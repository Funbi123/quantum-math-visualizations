```python
#Import necessary libraries

import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from scipy.special import factorial
from numpy.polynomial import Polynomial
```

```python
#Initialize the random number generator
rng = np.random.default_rng()

#Interactive slider
@interact(degree=(1, 100))

#Define the function that plots the graph
def polynomial(degree=3):
    #Specify the number of terms in the polynomual
    num_terms = np.arange(degree + 1)

    #Define the rule for generating the coefficients
    ran_coeff = 1/factorial(num_terms)

    #Generate the polynomials with the specified coordinates
    poly = Polynomial(ran_coeff)

    #specify the x and y values
    x_values = np.linspace(-5,5,100)
    y_values = poly(x_values)

    #Implemet the plot
    plt.figure(figsize=(8, 4))
    plt.plot(x_values, y_values, color="crimson", linewidth=2, label=f"Degree {degree}")
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.title(f"Taylor Polynomial Approximations Converging to $e^x$")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True, alpha=0.4)
    plt.legend()
    plt.show()
```
