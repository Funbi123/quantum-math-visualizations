```python
from IPython.display import display, Math 
#Now let's implement code 
#import necessary libraries 
import numpy as np
```

```python
#Define the basis vector of the quantum system 
state_1 = np.array([ 
    [0], 
    [1] 
]) 
state_2 = np.array([ 
    [1], 
    [0] 
])
```

```python
#Define complex coefficient/components of the basis vectors
alpha_1 = complex(input(r"Enter \alpha_1: ")) 
alpha_2 = complex(input(r"Enter \alpha_2: "))

#Generate a state vector 
psi = alpha_1 * state_1 + alpha_2 * state_2

#create a row vector of psi(bra_psi) to find the inner product(Conjugate Transpose)
bra_psi = np.conj(psi.T)

#Check if state vector is normalized witht the inner product 
is_normalized = bra_psi.dot(psi).item()

#Write the conditional statement to check if normalization condition is satisfied
if not np.isclose(is_normalized, 1.0): 
    norm_factor = 1/np.sqrt(is_normalized) 
    psi = norm_factor * psi 
else: 
    pass

bra_psi = np.conj(psi.T) 
print(f"The system is normalized: {bra_psi.dot(psi)}") 
print(f"The state vector is now given as: {psi}") 
```

```python
#Generate projection vector of the basis vectors 
p_0 = np.matmul(state_1, np.conj(state_1.T)) 
print(p_0)

p_1 = np.matmul(state_2, np.conj(state_2.T)) 
print(p_1)
```
