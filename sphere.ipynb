```python
def sphere(r):
#Define polar and azimuthal angles
    theta = np.linspace(0, (np.pi)/2, 50)
    phi = np.linspace(0, -np.pi, 50)
#Create a 2D grid of angles 
    theta, phi = np.meshgrid(theta, phi)

#Convert the spherical to rectangular coordinates
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)

#Setup figure and 3D axis
    fig = plt.figure(1, figsize=(12,4))
    ax = fig.add_subplot(111, projection='3d')

#turn off grid
    ax.axis('off')

    axis_length = r * 3

    #X-axis
    ax.plot([0, axis_length], [0,0], [0,0], color = 'red', linewidth = 2)
    ax.text(axis_length, 0, 0, 'X', color='red', fontsize=12, fontweight='bold')

    #Y-axis
    ax.plot([0,0],[0, axis_length], [0,0], color = 'green', linewidth = 2)
    ax.text(0, axis_length, 0, 'Y', color='green', fontsize=12, fontweight='bold')

    #Z-axis

    ax.plot([0,0],[0,0],[0, axis_length], color = 'blue', linewidth = 2)
    ax.text(0, 0, axis_length, 'Z', color='blue', fontsize=12, fontweight='bold')

#plot the 2D mesh arrays
    surface = ax.plot_surface(x, y, z, rstride =1, cstride =1, cmap = matplotlib.cm.coolwarm, linewidth=0)
    ax.set_box_aspect([1,1,1])

    plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

r = 5

theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)

theta, phi = np.meshgrid(theta, phi)

x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)

fig = go.Figure(
    data=[
        go.Surface(
            x=x,
            y=y,
            z=z,
            colorscale='Viridis'
        )
    ]
)

fig.update_layout(
    scene=dict(
        aspectmode='data'
    )
)

fig.show()
```
