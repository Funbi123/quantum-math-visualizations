A quantum state is defined by the superposition of the possible basis vector states multiplied by complex amplitudes. This state contains all the information about the physical system it describes and these information can be extracted as probabilities. This project achieves the following aims:

1. Represent a quantum state in the form of its basis vector
2. Normalize the state vector to achieve the probability distribution condition.
3. Determine the projection matrix of the state and understand it significance
4. Rotate the basis and change it to see its effect on the system
6. Visualize it on a Bloch Sphere

I would do all of these first with a qubit and then the real project would be on photon polarization.
